```python
import os
import glob
import pypandoc
from pathlib import Path
```

```python
input_dir = "/Users/abbyj/gvp/text"
```

```python
print("Exists?", Path(input_dir).exists())
```

```python
print(list((Path(input_dir).glob("*.docx"))))
```

```python
for filepath in glob.glob(os.path.join(input_dir, "*.docx")):
    output_path = filepath.replace(".docx", ".txt")
    
    text = pypandoc.convert_file(filepath, 'plain')
    
    with open(output_path, "w", encoding="utf-8") as f:
        f.write(text)
```

```python
data_dir = Path(input_dir)

for file in data_dir.glob("*.docx"):
    file.unlink()  # deletes the file
    
```
